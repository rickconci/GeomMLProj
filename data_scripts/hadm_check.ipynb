```python
import pandas as pd, os, re

input_pkl_physio = '/Users/riccardoconci/Local_documents/!!GeomML_2025/GeomMLProj/temp_dfs/sorted_filtered_df.pkl'
input_pkl_base = '/Users/riccardoconci/Local_documents/!!GeomML_2025/GeomMLProj/temp_dfs/merged_with_disch_df_final_filtered.pkl'
cache_dir = '/Users/riccardoconci/Local_documents/!!GeomML_2025/GeomMLProj/temp_dfs/precomputed_tensors'



# 1. All hadm_ids from the input DataFrame
df = pd.read_pickle(input_pkl_base)
all_ids = set(df['hadm_id'].unique())

# 2. All hadm_ids for which a .pt file exists
cached_files = os.listdir(cache_dir)
cached_ids = set()
for fn in cached_files:
    match = re.search(r'tensor_cache_(\d+)\.pt', fn)
    if match:
        cached_ids.add(int(match.group(1)))

        
missing = all_ids - cached_ids
print(f"{len(missing)} hadm_ids never got a .pt:", missing)
```
